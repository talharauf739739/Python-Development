**Pagination**

```python
name = []
title = []
email = []

# BELOW CODE FOR PAGINATION... FOR MULTIPLE PAGES

for i in range (1,84):
    
    # website in variable
    #https://www.smfm.org/members/search
    #https://www.smfm.org/members/search?page=2
    website = 'https://deptmedicine.arizona.edu/directory?page='+str(i)
    #website = 'https://www.smfm.org/members/search?page='+str(i)
    
    # request to website
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('div', {'class' : 'peoplewrapper'})
    
    # loop through results
    

  
    for result in results:
      # Name
      try:
        name.append(result.find('a').get_text()) 
      except:
        name.append('Doctor')
    
 
      # Title
      try:
        title.append(result.find('em').get_text())
      except:
        title.append('n/a')
    
      # Email
      try:
        Email = result.find_all('div')

        email.append(Email[3].find('a').get_text())
      except:
        email.append('n/a')
        
    
```

```python
# dictionary
Hey_Doctors = pd.DataFrame({'Name': name, 'Title':title, 'Email':email})
```
