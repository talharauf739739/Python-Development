```python
from time import sleep

def scrolling():
    for i in range(7):
        pre_scroll = driver.find_element(By.CSS_SELECTOR, 'html')
        #pre_scroll = driver.find_element(By.TAG_NAME, 'body')
        #pre_scroll.click()
        new_scroll = pre_scroll.send_keys(Keys.PAGE_DOWN)
        #new_scroll =driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
        
        time.sleep(2)
        while True:
            new_scroll
            time.sleep(2)
            if new_scroll == pre_scroll:
                break
            pre_scroll = new_scroll
```
